```python
import cv2 # OpenCV
import numpy as np
from matplotlib import pyplot as plt
import tensorflow as tf
import multiprocessing
import os
import zipfile
import shutil
```

```python
num_cores = multiprocessing.cpu_count()
config = tf.compat.v1.ConfigProto(inter_op_parallelism_threads=num_cores,
                                  intra_op_parallelism_threads=num_cores)
tf.compat.v1.Session(config=config)
```

```python
diretorio_pai = '/home/cristiano/Imagens/Imagens_Satelites/SENTINEL-2/TREINO/'  # Substitua pelo caminho do diretório pai contendo as pastas compactadas

# Listar os arquivos no diretório pai
arquivos = os.listdir(diretorio_pai)

# Iterar sobre os arquivos
for arquivo in arquivos:
    caminho_arquivo = os.path.join(diretorio_pai, arquivo)  # Obter o caminho completo do arquivo

    if arquivo.endswith('.zip'):  # Verificar se é um arquivo zip
        nome_pasta = os.path.splitext(arquivo)[0]  # Obter o nome da pasta (sem a extensão .zip)

        # Criar o diretório para descompactar
        diretorio_descompactado = os.path.join(diretorio_pai, nome_pasta)

        # Descompactar o arquivo zip
        with zipfile.ZipFile(caminho_arquivo, 'r') as zip_ref:
            zip_ref.extractall(diretorio_descompactado)

        print(f'Pasta descompactada: {nome_pasta}')
```

```python

# Listar as pastas no diretório pai
pastas = os.listdir(diretorio_pai)

# Iterar sobre as pastas
for pasta in pastas:
    caminho_pasta = os.path.join(diretorio_pai, pasta)  # Obter o caminho completo da pasta

    if os.path.isdir(caminho_pasta):  # Verificar se é um diretório
        # Criar o diretório "images"
        diretorio_images = os.path.join(caminho_pasta, 'images')
        os.makedirs(diretorio_images, exist_ok=True)

        # Criar o diretório "mask"
        diretorio_mask = os.path.join(caminho_pasta, 'mask')
        os.makedirs(diretorio_mask, exist_ok=True)

        print(f'Diretórios criados em {pasta}: {diretorio_images}, {diretorio_mask}')
```

```python
diretorio_treino = '/home/cristiano/Imagens/Imagens_Satelites/SENTINEL-2/TREINO/'  # Substitua pelo caminho do diretório "treino"

# Iterar sobre os diretórios dentro de "treino"
for diretorio in os.listdir(diretorio_treino):
    caminho_diretorio = os.path.join(diretorio_treino, diretorio)  # Obter o caminho completo do diretório
    
    if os.path.isdir(caminho_diretorio):  # Verificar se é um diretório
        # Mover os arquivos .tiff para o diretório "images" dentro do diretório atual
        for arquivo in os.listdir(caminho_diretorio):
            caminho_arquivo = os.path.join(caminho_diretorio, arquivo)  # Obter o caminho completo do arquivo
            
            if arquivo.lower().endswith('.tiff'):  # Verificar se é um arquivo .tiff
                destino = os.path.join(caminho_diretorio, 'images', arquivo)  # Obter o caminho de destino
                shutil.move(caminho_arquivo, destino)  # Mover o arquivo para o diretório de destino

                print(f'Arquivo {arquivo} movido para {destino}')
```
